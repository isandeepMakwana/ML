```python
import sklearn.datasets
import sklearn.svm
import matplotlib.pyplot
import numpy
```

```python
# creating sample data
x, y = sklearn.datasets.make_blobs(n_samples=2, centers=2, n_features=2,random_state=20)

# query data
queryData1 = [[x[0][0],x[0][1]]]
queryData2 = [[x[1][0] , x[1][1]]]
print(queryData1)
print(queryData2)

# svm 
classifier = sklearn.svm.SVC(kernel="linear" , C=500)
classifier.fit(x,y)
print(classifier.predict(queryData1))
print(classifier.predict(queryData2))
```

```python
# plotting
matplotlib.pyplot.scatter(x[:, 0],x[:, 1] , c=y)
axes=matplotlib.pyplot.gca()
xlimit = axes.get_xlim()
ylimit = axes.get_ylim()
print("xlimits, ylimits : ", xlimit, ylimit)

xx = numpy.linspace(xlimit[0],xlimit[1],5)
yy = numpy.linspace(ylimit[0],ylimit[1],5)
print("xx : ",xx)
print("yy : ",yy)
xxx ,yyy = numpy.meshgrid(xx,yy)
print("xxx : ", xxx)
print("yyy : ", yyy)
# flatten
xxx = xxx.ravel()
yyy = yyy.ravel()

print("*"*50)
print(type(xxx))

print("xxx : ",xxx,"\nyyy : ",yyy)
xxxyyy = numpy.c_[xxx,yyy]
print("xxxyyy : ",xxxyyy)
print(xxxyyy.shape)

zz = classifier.decision_function(xxxyyy)
print("zz : ",zz)
print(xx.shape)
zz = zz.reshape(5,5)
print("reshape : ",zz)

# conture graph
matplotlib.pyplot.contour(xx,yy,zz , levels=[-1,0,1] , linestyles=["--", "-", "--"])

print(classifier.support_vectors_)
# a=[[2.12015254 5.61101265],[6.745782   6.87644584]]
for i in range(len(classifier.support_vectors_)):
    matplotlib.pyplot.scatter(classifier.support_vectors_[i][0],classifier.support_vectors_[i][1],c="red",s=100,edgecolors='k',linewidths=2,facecolors='none')
matplotlib.pyplot.show()
```
