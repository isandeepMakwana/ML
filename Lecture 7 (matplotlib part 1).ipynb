# #Another Example :
# Sales Predictions for year 2021 / 2022


````python
sales2021 = numpy.random.randint(500000,size=12)
sales2022 = numpy.random.randint(500000,size=12)
print('sales2021\n',sales2021)
print('sales2022\n',sales2022)
matplotlib.pyplot.plot(sales2021,color='r')
matplotlib.pyplot.plot(sales2022,color='g')
matplotlib.pyplot.xlabel("Months")
matplotlib.pyplot.ylabel("Sales")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.title("Sales Predictions for Year 2021 / 2022")
matplotlib.pyplot.show()
````

````python
# identify which line is sales21 and 22
sales2021 = numpy.random.randint(500000,size=12)
sales2022 = numpy.random.randint(500000,size=12)
print('sales2021\n',sales2021)
print('sales2022\n',sales2022)
matplotlib.pyplot.plot(sales2021,color='r',label="2021")
matplotlib.pyplot.plot(sales2022,color='g',label="2022")
matplotlib.pyplot.xlabel("Months")
matplotlib.pyplot.ylabel("Sales")
matplotlib.pyplot.grid(True)
matplotlib.pyplot.legend()
matplotlib.pyplot.xticks(range(12),range(1,13))
matplotlib.pyplot.title("Sales Predictions for Year 2021 / 2022")
matplotlib.pyplot.show()
````

````python
# more opetiosn
sales2021 = numpy.random.randint(500000,size=12)
sales2022 = numpy.random.randint(500000,size=12)
print('sales2021\n',sales2021)
print('sales2022\n',sales2022)
matplotlib.pyplot.plot(sales2021,color='r',label="2021")
matplotlib.pyplot.plot(sales2022,color='g',label="2022")
matplotlib.pyplot.xlabel("Months")
matplotlib.pyplot.ylabel("Sales")
matplotlib.pyplot.grid(True)
# matplotlib.pyplot.legend()
# matplotlib.pyplot.legend(loc="best")
matplotlib.pyplot.legend(loc="upper left")
matplotlib.pyplot.xticks(range(12),range(1,13))
matplotlib.pyplot.title("Sales Predictions for Year 2021 / 2022")
matplotlib.pyplot.show()
````
