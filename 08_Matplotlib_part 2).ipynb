# the way of writing linestyle 

```python
a=[10,20,30,40]
b=[80,160,200,500]
# 1,2,3,4,s,p,h,<,>,v,^.*,+,
matplotlib.pyplot.plot(a,b,linestyle=':',color="#11aabb") # --,-.,:,-1
matplotlib.pyplot.savefig("abc2.png")
matplotlib.pyplot.show()
```

# historgram


```python
import matplotlib.pyplot
import numpy
a=numpy.random.randint(20,100,size=15)
print(a)
matplotlib.pyplot.hist(a,15)
matplotlib.pyplot.show()
```

```python
a=(10,15,25,50)
captions =("Pizza","Burger","Coke","Finger chips")
matplotlib.pyplot.pie(a,labels=captions)
matplotlib.pyplot.show()
```

```python
a=(10,15,25,50)
captions =("Pizza","Burger","Coke","Finger chips")
matplotlib.pyplot.pie(a,labels=captions,explode=(.1,.1,.1,.4))
matplotlib.pyplot.show()
```
