```python
import sklearn.datasets
import sklearn.svm
import matplotlib.pyplot
```

```python
x, y = sklearn.datasets.make_blobs(
    n_samples = 100,
    centers=2,
    n_features=2,
    random_state=20
)
print(x)
print(y)
```

```python
classifier = sklearn.svm.SVC(kernel = 'linear',C=500) # C=500 is for regularization
print(type(classifier))
classifier.fit(x,y)
```

```python
queryData1 =[[x[1][0] , x[1][1]]]
queryData2 =[[x[73][0] , x[73][1]]]
print(queryData1)
print(queryData2)
```

```python
# targets
print(y[1])
print(y[73])
```

```python
print(classifier.predict(queryData1))
print(classifier.predict(queryData2))
```
