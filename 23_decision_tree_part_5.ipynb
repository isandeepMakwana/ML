```python
# spcify the Arrow
import matplotlib.pyplot
matplotlib.pyplot.scatter(10,2,color="r")
matplotlib.pyplot.annotate('good',(10,2),xytext=(20,20) , textcoords='offset points' ,ha='center' , arrowprops=dict(facecolor='green'))
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
matplotlib.pyplot.scatter(10,2,color="r")
matplotlib.pyplot.annotate('good',(10,2),xytext=(80,80) , textcoords='offset points' ,ha='center' , arrowprops=dict(facecolor='green'))
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot
matplotlib.pyplot.scatter(10,2,color="r")
matplotlib.pyplot.annotate('good',(10,2),xytext=(80,80) , textcoords='offset points' ,ha='center' , arrowprops=dict(facecolor='green' , shrink=0.1))
matplotlib.pyplot.show()
```

# subplot

```python
# if we have one row and so don't do duble subscripting 
import matplotlib.pyplot
f, m = matplotlib.pyplot.subplots(1,2) # look like matrix 1 row , 2 col
a=((10,20),(30,40),(60,50))
m[0].plot(a)
m[1].plot(a)
matplotlib.pyplot.show()
# we can pass different plots as well
```

```python

import matplotlib.pyplot
f, m = matplotlib.pyplot.subplots(2,2)
a=((10,20),(30,40),(60,50))
m[0][0].plot(a)
m[0][1].plot(a)
b=((100,200),(300,400),(600,500))
m[1][0].plot(b)
m[1][1].scatter(10,2,color='r')
matplotlib.pyplot.show()
# we can pass different plots as well
```
