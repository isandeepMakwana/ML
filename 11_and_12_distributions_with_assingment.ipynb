
# Normal Distribution 

```python
from pprint import pprint as print
import numpy 
import matplotlib.pyplot

data = numpy.random.normal(size=50)
print(data)
matplotlib.pyplot.hist(data, bins=5)
matplotlib.pyplot.show()
```

```python
import numpy 
import matplotlib.pyplot

data = numpy.random.normal(size=500)
print(data)
matplotlib.pyplot.hist(data, bins=10)
matplotlib.pyplot.show()
```

# mu and sigma 

```python
import numpy 
import matplotlib.pyplot
mu = 0 # mean
sigma = 0.1 # Standard deviation
data = numpy.random.normal(mu, sigma,size=50)
print(data)
matplotlib.pyplot.hist(data, bins=5)
matplotlib.pyplot.show()
```

# Uniform Distribution

```python

import numpy 
import matplotlib.pyplot

data = numpy.random.uniform(size=50)
print(data)
matplotlib.pyplot.hist(data, bins=5)
matplotlib.pyplot.show()
```

```python
import numpy 
import matplotlib.pyplot

data = numpy.random.uniform(size=500)
print(data)
matplotlib.pyplot.hist(data, bins=10)
matplotlib.pyplot.show()
```

# Uniform distribution using scipy

```python
import scipy.stats
import matplotlib.pyplot

data = scipy.stats.uniform.rvs(size=500,loc=10,scale =50)
print(data)
matplotlib.pyplot.hist(data, bins=10)
matplotlib.pyplot.show()
```

# Assingments
- find MAE(Mean AbsoluteError)
- find MSE(Mean Squared Error)
- find RMSE (root mean squared error)

```python
#MAE
import numpy
a=[100,200,300,100,200,300,400,500,300,700,800,900,1000,800]
p=[100,150,300,100,199,300,400,450,300,700,800,800,900,1000]


# d=numpy.subtract(a,p)
# print(d)
diff_list = [abs(a[i]-p[i]) for i in range(len(a))]
print(diff_list)
mae=numpy.sum(diff_list)/len(a)
print(mae)
```

```python
# mse
import numpy
a=[2, 3, 5, 5, 9]
p=[3, 3, 8, 7, 6]
diff_list = numpy.subtract(a,p)
squred_diff=numpy.square(diff_list)
mse = numpy.sum(squred_diff)/len(a)
print(mse)
```

```python
# rmse
import numpy
a=[2, 3, 5, 5, 9]
p=[3, 3, 8, 7, 6]
diff_list = numpy.subtract(a,p)
squred_diff=numpy.square(diff_list)
mse = numpy.sum(squred_diff)/len(a)
print(numpy.sqrt(mse))
```
