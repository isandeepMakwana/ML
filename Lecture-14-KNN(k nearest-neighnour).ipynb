# KNN(K-Nearest-neighour) ALGO

## setps
<ol>
<li>data points (data sets)</li>
<li>Decide the value of K (int sqrt(len of datasets))</li>
<li>Query data | new data  , whose calsee is to be identified point</li>
<li>Calculate the distance between query point and each element in data set and store in to a like ('A',25)</li>
<li>Sort the list by the values</li>
<li>Choose the top k Rows</li>
<li>Count the frequency of A and B top K Rows</li>
<li>The identified class is the one most frequency count (MODE)</li>
</ol>


```python
import statistics
import math
import matplotlib.pyplot
import numpy

"""
#following Represents  A Class
matplotlib.pyplot.plot(10,20,"ro")
matplotlib.pyplot.plot(20,25,"ro")
matplotlib.pyplot.plot(15,20,"ro")
matplotlib.pyplot.plot(30,45,"ro")
matplotlib.pyplot.plot(10,10,"ro")
matplotlib.pyplot.plot(50,25,"ro")
matplotlib.pyplot.plot(30,55,"ro")
matplotlib.pyplot.plot(40,60,"ro")
#following Represents  B Class
matplotlib.pyplot.plot(100,120,"go")
matplotlib.pyplot.plot(120,125,"go")
matplotlib.pyplot.plot(115,120,"go")
matplotlib.pyplot.plot(130,145,"go")
matplotlib.pyplot.plot(100,100,"go")
matplotlib.pyplot.plot(150,125,"go")
matplotlib.pyplot.plot(130,155,"go")
matplotlib.pyplot.plot(140,160,"go")


# create a Black
queryX = 80
queryY = 100
matplotlib.pyplot.plot(queryX , queryY ,color="b" , marker ="o" , markersize=20)
"""
# KNN - ALGO - CODE


a = numpy.array(
    [
        [10, 20], 
        [20, 25], 
        [15, 20], 
        [30, 45], 
        [10, 10], 
        [50, 25], 
        [30, 55], 
        [40, 60],
    ]
)
b = numpy.array(
    [
        [100, 120],
        [120, 125],
        [115, 120],
        [130, 145],
        [100, 100],
        [150, 125],
        [130, 155],
        [140, 160],
    ]
)

# but we need all the x an y on onle two [[],[]]

# what we can do transpose a and b

x, y = a.T
print(x)
matplotlib.pyplot.scatter(x, y, color="r")
x, y = b.T
print(y)
matplotlib.pyplot.scatter(x, y, color="g")

matplotlib.pyplot.grid(True)
queryX = 60
queryY = 100
matplotlib.pyplot.plot(queryX,queryY ,color='b',marker='o',markersize=20)

# matplotlib.pyplot.show()

# KNN Alogrithm starts hear


distanceList = []
for p in a:
    xd = abs(p[0] - queryX)
    yd = abs(p[1] - queryY)
    d = math.sqrt(xd**2 + yd**2)
    distanceList.append(("A", d))


for p in b:
    xd = abs(p[0] - queryX)
    yd = abs(p[1] - queryY)
    d = math.sqrt(xd**2 + yd**2)
    distanceList.append(("B", d))


print("*" * 50)
print(distanceList)

distanceList.sort(key=lambda x: x[1])

print("*" * 50)
kFactor = int(math.sqrt(len(a) + len(b)))

if kFactor % 2 == 0:
    kFactor += 1
print("kFactor : ", kFactor)
kElementFromTop = distanceList[:kFactor]
print(kElementFromTop)
classes = [x[0] for x in kElementFromTop]
print(classes)
identifiedClass = statistics.mode(classes)
print(f"{queryX} , {queryY} belongs to {identifiedClass}")

```

# Assignment KNN with other exmaple (Classification)

```python
# data prep
import pandas as pd

df=pd.read_csv('data/MOCK_DATA.csv')
maths = df['maths']
hindi = df['hindi']
english=df['english']
computer=df['computer']

student_lst = [
    (maths[i], hindi[i], english[i], computer[i]) for i in range(len(maths))
]
student_lst.sort(key=lambda x : x[0]-x[1]-x[2]-x[3])

excellent_students=[]
average_students=[]
ok_students=[]
for k, i in enumerate(student_lst):
    if i[0] < 33 or i[1]<33 or i[2]<33 or i[3]<33:
        ok_students.append(i)
    elif k<17:
        excellent_students.append(i)
    elif k<34:
        average_students.append(i)
    else:
        ok_students.append(i)
import math
print(f"there was three groups 1. excellent_students = {excellent_students} \n 2. average_students = {average_students}\n 3. ok_students = {ok_students} ")
# Query data 
new_student = (56,82,51,40) # i think it is excellent_student 
# find the student_group
import statistics
def find_group(p,q ,group):
    a=abs(p[0]-q[0])
    b=abs(p[1]-q[1])
    c=abs(p[2]-q[2])
    d=abs(p[3]-q[3])
    return (group,a+b+c+d)

lst = [find_group(i,new_student,'excellence') for i in excellent_students]
lst.extend(find_group(i,new_student,'average') for i in average_students)
lst.extend(find_group(i,new_student,'ok') for i in ok_students)

k = int(math.sqrt(len(excellent_students)+len(average_students)+len(ok_students)))
if k%3==0:
    k+=1
print(k)

lst.sort(key=lambda r : r[1])
ktopRows = lst[:k]
print(ktopRows)
print(statistics.mode([ i[0] for i in ktopRows]))
```
